```python
#Entrypoint 2.x
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Spark SQL basic example").enableHiveSupport().getOrCreate()

# On yarn:
# spark = SparkSession.builder.appName("Spark SQL basic example").enableHiveSupport().master("yarn").getOrCreate()
# specify .master("yarn")

sc = spark.sparkContext
```

```python
# Read raw events from local filesystem
events_df = spark.read \
    .option("header", "false") \
    .csv("file:///home/talentum/test-jupyter/ecommerce-clickstream-etl/data/raw/events.csv") \
    .toDF("timestamp", "visitor_id", "event_type", "item_id", "transaction_id")

# Basic sanity check
events_df.show(5)
print(events_df.count())
# ðŸ”´ VERY IMPORTANT:
# Persist this dataframe so other notebooks can use it
events_df.write.mode("overwrite").parquet(
    "file:///home/talentum/test-jupyter/ecommerce-clickstream-etl/data/processed/events_base"
)
```
